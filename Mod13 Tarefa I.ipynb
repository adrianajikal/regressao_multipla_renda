```python
import pandas as pd
import numpy as np
import patsy
import statsmodels.api as sm
```

```python
df = pd.read_csv('previsao_de_renda.csv')
df.head()
```

```python
df.info()
```

1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.

2. Remova a variável menos significante e analise:
    - Observe os indicadores que vimos, e avalie se o modelo melhorou ou piorou na sua opinião.
    - Observe os parâmetros e veja se algum se alterou muito.

3. Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 
    

```python
#1. Modelo patsy para prever log(renda) considerando todas as variáveis:
y, X = patsy.dmatrices('np.log(renda) ~  sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda + educacao + estado_civil + tipo_residencia + idade + tempo_emprego + qt_pessoas_residencia', data = df)
sm.OLS(y,X).fit().summary()
```

As variaveis com maiores valores p (acima de 0,05) são:
- tipo_renda[T.Bolsista] : 0.360
- tipo_renda[T.Pensionista] : 0.201
- educacao[T.Pós graduação] : 0.501
- educacao[T.Secundário] : 0.844
- educacao[T.Superior completo] : 0.194
- educacao[T.Superior incompleto] : 0.579
- estado_civil[T.União] : 0.174
- tipo_residencia[T.Casa] : 0.670
- tipo_residencia[T.Estúdio] : 0.503
- tipo_residencia[T.Governamental] : 0.387

```python
#2)Remoçao da variável menos significante: educaçao
y, X = patsy.dmatrices('np.log(renda) ~  sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda + estado_civil + tipo_residencia + idade + tempo_emprego + qt_pessoas_residencia', data = df)
sm.OLS(y,X).fit().summary()
```

**Observaçoes:**
- R²: diminuiu de 0.357 para 0.354 
- R² ajustado: diminuiu de 0.356 para 0.353
- AIC: aumentou de 2.719e+04 para 2.724e+04
- BIC: aumentou de 2.737e+04 para 2.740e+04

**O modelo piorou um pouco conforme os dados acima porem os coeficientes principais se mantiveram estaveis sugerindo que a variavel educacao removida nao foi muito relevante.**

```python
#3. Remoçao das demais variaveis com valor p  acima de 5%
y, X = patsy.dmatrices('np.log(renda) ~  sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + estado_civil + idade + tempo_emprego + qt_pessoas_residencia', data = df)
sm.OLS(y,X).fit().summary()
```

**Observaçoes:**
- R²: diminuiu de 0.357 para 0.347 
- R² ajustado: diminuiu de 0.356 para 0.347
- AIC: aumentou de 2.719e+04 para 2.736e+04
- BIC: aumentou de 2.737e+04 para 2.745e+04

**O modelo ficou pior ajustado em relação aos anteriores, mesmo tirando as variaveis com valor p acima de 5% .**
